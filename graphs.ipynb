```python
C = nx.Graph()
```

```python
n = 4
amigos = np.arange(1,n+1)
amigos = ['amigo_'+str(i) for i in amigos]
amigos
```

```python
C.add_nodes_from(amigos)
```

```python
plt.figure(figsize=(5,5))
nx.draw(C, with_labels = True, node_size=3000)
#plt.tight_layout()
plt.show()
```

```python
muchos = combinations(amigos, 2)
muchos = list(muchos)
muchos[:5]
```

```python
C.add_edges_from(muchos)
```

```python
plt.figure(figsize=(5,5))
print(f'Densidad:{nx.density(C)}')
nx.draw(C, with_labels = True, node_size=3000)
plt.show()
```

```python
A = nx.adjacency_matrix(C)
B = A.todense()
B
```

```python
print(B.shape)
plt.imshow(B)
```
