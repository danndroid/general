```python
class Node:
    length = 0
    convex_subsequence = []


def get_longest_convex_subsequence_length(integer_list):
    
    M = np.zeros((5,5), dtype=int) #####

    if len(integer_list) > 1:
        m = [[Node() for i in range(len(integer_list))] for j in range(len(integer_list))]
        #print('m=',len(m))
        i = len(integer_list) - 2
        #print('i=',i)
        while i >= 0:
            j = len(integer_list) - 1
            print('(i,j)=',(i,j))
            while j >= i + 1:
                print('(j)',j, '>=', i + 1,'(i+1)')
                m[i][j].length = 2
                M[i][j] = m[i][j].length #####
                print('J=',j)
                print(M) #####
                arr = [integer_list[i], integer_list[j]]
                #print(arr)
                m[i][j].convex_subsequence = arr
                k = j + 1
                print('k<-',k)
                while k < len(integer_list):
                    print('(',i,j,k,')')
                    print(integer_list[i], integer_list[j],integer_list[k])
                    if integer_list[k] - integer_list[j] > integer_list[j] - integer_list[i]:
                        print(integer_list[k], integer_list[j], '>', integer_list[j], integer_list[i])
                        if m[i][j].length < (1 + m[j][k].length):
                            print('(m[i,j])',m[i][j].length, '<' ,'1', '+', m[j][k].length,'(m[j,k])')
                            m[i][j].convex_subsequence = [integer_list[i]] + m[j][k].convex_subsequence
                        m[i][j].length = max(m[i][j].length, 1 + m[j][k].length)
                        M[i][j] = m[i][j].length #####
                        print('K=',k)
                        print(M) #####
                    print('EX')
                    k += 1
                    print('K=',j)
                j -= 1
                print('J=',j)
            i -= 1
            print('I=',j)
            #print('(i,j)=',(i,j),'\n')
            #print(m[0][1].convex_subsequence)
            
        return m, m[0][1]
    else:
        node = Node()
        node.length = len(integer_list)
        node.convex_subsequence = integer_list
        return node
```
