```python
plt.title('Cumulative Risk Profile',fontsize=20)

x = [4, 8, 8]
y = [.73, .73, 1]
#plt.scatter(x=x,y=y, color='b')
plt.plot([4,4],[0,.73], color='blue') # (X1,X2)(Y1,Y2)
plt.plot([4,8],[.73,.73], color='blue')
plt.plot([8,8],[.73,1], color='blue', label='A:A1')

x = np.array([0, 4.5, 4.5, 15, 15])
y = [0.365, 0.365, 0.865, 0.865, 1]
#plt.scatter(x,y, color='g')
plt.plot([0,0],[0, 0.365], color='g') # (X1,X2)(Y1,Y2)
plt.plot([0,4.5],[0.365, 0.365], color='g')
plt.plot([4.5, 4.5],[0.365, 0.865], color='g')
plt.plot([4.5, 15],[0.865, 0.865], color='g')
plt.plot([15, 15],[0.865, 1], color='g',label='A:A2')

x = [0, 10, 10]
y = [.45, .45, .55]
#plt.bar(x,y, label='B')
plt.plot([0,0],[0,.45], color='r') # (X1,X2)(Y1,Y2)
plt.plot([0,10],[.45,.45], color='r')
plt.plot([10,10],[.45, 1], color='r', label='B')

plt.xlabel('Outcome x', fontsize=15)
plt.ylabel('P(x)',fontsize=15)
plt.xlim([0,16])
#plt.ylim([0,1])
plt.legend()
plt.show()
```
