```python
x = [48.83, 57.17, 73.83, 40.50, 65.5]
y = [0.09+0.28, 0.42, 0.09, 0.06, 0.06]
print(np.sum(y))
plt.title('Madison',fontsize=20)
plt.bar(x,y)
#plt.bar(x,y, alpha)
plt.xlabel('Score x', fontsize=15)
plt.ylabel('P(x)',fontsize=15)
plt.ylim([0,1])
plt.show()
```

```python
x = [29.17, 35.83, 43.33]
y = [.15, .7, .15]

plt.title('MPR',fontsize=20)
plt.bar(x,y)
plt.xlabel('Score x', fontsize=15)
plt.ylabel('P(x)',fontsize=15)
plt.ylim([0,1])
plt.show()
```

```python
x = [50]
y = [1]

plt.title('P-Pizza',fontsize=20)
plt.bar(x,y)
plt.xlabel('Score x', fontsize=15)
plt.ylabel('P(x)',fontsize=15)
plt.ylim([0,1])
plt.xlim([40,55])
plt.show()
```

```python
x = [48.83, 57.17, 73.83, 40.50, 65.5]
y = [0.09+0.28, 0.42, 0.09, 0.06, 0.06]
y = [0.06, 0.09+0.28, 0.42, 0.09, 0.06]
y = np.cumsum(y)
print(y)
y = [.15, .7, .15]
y = np.cumsum(y)
print(y)
y = [1]
y = np.cumsum(y)
print(y)
```

```python
plt.title('Cumulative Risk Profile',fontsize=20)


plt.plot([40.50,40.50],[0, 0.06], color='blue') # (X1,X2)(Y1,Y2)
plt.plot([40.5,48.83],[.06,0.06], color='blue')
plt.plot([48.83, 48.83],[.06,0.43], color='blue')
plt.plot([48.83, 57.17],[.43,.43], color='blue')
plt.plot([57.17, 57.17],[.43,0.85], color='blue')
plt.plot([57.17, 65.50],[0.85,0.85], color='blue')
plt.plot([65.50, 65.50],[0.85,0.94], color='blue')
plt.plot([65.50, 73.83],[0.94,0.94], color='blue')
plt.plot([73.83, 73.83],[0.94,1], color='blue', label='Madison')

plt.plot([29.17,29.17],[0, 0.15], color='g') # (X1,X2)(Y1,Y2)
plt.plot([29.17,35.83],[0.15, 0.15], color='g') # (X1,X2)(Y1,Y2)
plt.plot([35.83,35.83],[0.15, 0.85], color='g') # (X1,X2)(Y1,Y2)
plt.plot([35.83,43.33],[0.85, 0.85], color='g') # (X1,X2)(Y1,Y2)
plt.plot([43.33,43.33],[0.85, 1], color='g',label='MPR') # (X1,X2)(Y1,Y2)

plt.plot([50,50],[0, 1], color='r',label='P-Pizza') # (X1,X2)(Y1,Y2)


plt.xlabel('Score x', fontsize=15)
plt.ylabel('P(x)',fontsize=15)
#plt.xlim([0,16])
#plt.ylim([0,1])
plt.legend()
plt.show()
```
