```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dj = [31.5,31.6,31.7,31.8,31.9, 32, 32.1,32.2,32.3,32.4,32.5]
fs = [1,2,3,4,5,6,4,2,2,1,1]
qt = [0.05, 0.25, 0.50, 0.75, 0.95]
```

```python
df = pd.DataFrame({'djia':dj,
             'vote':fs})
df['cumulative'] = df.vote.cumsum()
df['cdf'] = df.cumulative / df.vote.sum()
df
```

```python
fig , axes = plt.subplots(1,3, figsize=(17,5))

plt.suptitle('DJIA CDF assessment', fontsize=20)
ax = axes[0]
sns.barplot(x=df.djia, y=df.vote, color='gray', ax=ax)
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('Estimation', fontsize=15)

ax = axes[1]
sns.lineplot(x=df.djia, y=df.cdf, color='gray', ls='--', lw=3, ax=ax)
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('CFD estimation', fontsize=15)

ax = axes[2]
sns.lineplot(x=df.djia, y=df.cdf, color='gray', ls='--', lw=3, ax=ax)
for f in qt:
    value = df[df['cdf']<=f]['djia'].max()
    plt.plot([value], [f], marker='o', markersize=7, label=f'frac({f})={value}')
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('CFD estimation', fontsize=15)


plt.legend()
plt.show()
```

```python
t_probs = [0, 0.185, 0.630, 0.185 , 0.0]
t_values = []
for f in [0, 0.05, 0.50, 0.95, 1]:
    if f == 0:
        t_values.append(df['djia'].min())
    else:
        value = df[df['cdf']<=f]['djia'].max()
        t_values.append(value)
df2 = pd.DataFrame({'frac_t':[0, 0.05, 0.50, 0.95, 1],
                    'prob_t':t_probs,
                   'value_t':t_values})
df2['cdf_t'] = df2.prob_t.cumsum()
df2['value_t2'] = df2['value_t'].shift(-1).fillna(33)
df2
```

```python
s_probs = [0, 0.167, 0.666, 0.167, 0]
t_values = []
for f in [0, 0.10, 0.50, 0.90, 1]:
    if f == 0:
        t_values.append(df['djia'].min())
    else:
        value = df[df['cdf']<=f]['djia'].max()
        t_values.append(value)
df2['frac_s'] = s_probs
df2['prob_s'] = s_probs
df2['value_s'] = t_values
df2['cdf_s'] = df2.prob_s.cumsum()
df2['value_s2'] = df2['value_s'].shift(-1).fillna(33)
df2
```

```python
fig , axes = plt.subplots(1,3, figsize=(17,5))

plt.suptitle('DJIA CDF approximation', fontsize=20)
ax = axes[0]
sns.lineplot(x=df.djia, y=df.cdf, color='gray', ls='--', lw=3, ax=ax)
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('CFD estimation', fontsize=15)

ax = axes[1]
sns.lineplot(x=df.djia, y=df.cdf, color='gray', ls='--', lw=3, ax=ax)
ax.step(x=df2.value_t2, y=df2.cdf_t, color='b', ls='--', lw=3, label='Pearson-Tuckey')
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('CFD estimation', fontsize=15)
ax.legend()

ax = axes[2]
sns.lineplot(x=df.djia, y=df.cdf, color='gray', ls='--', lw=3, ax=ax)
ax.step(x=df2.value_s2, y=df2.cdf_s, color='b', ls='--', lw=3, label='Swanson-Megill')
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('DJIA value (thousands)', fontsize=15)
ax.set_ylabel('CFD estimation', fontsize=15)


plt.legend()
plt.show()
```
